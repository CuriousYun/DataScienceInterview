## Question 41 - Rolling dice to win...again

Suppose you are playing a game where there are two fair six-sided dice, and every time you roll the dice and they add up to 9, you win $50. 

However, to roll the dice costs $20 to play. Is this a game you're willing to play? 

You should be able to attach an expected value to each dice roll using probability theory here.

```python
import pandas as pd
#A dice has numbers from 1 to 6 and each face has 1/6 probability.
dice=pd.DataFrame({'number':[1,2,3,4,5,6],'probability':[1/6]*6})
dice
```

```python
#if we have two dices, what will happen?
twodice=[]
for i in dice['number']:
    for j in dice['number']:
        twodice.append([i,j,1/6*1/6])
result = pd.DataFrame(twodice,columns=['first','second','probability'])
```

```python
result.head()
```

```python
#our return is based on the sum value of faces of two dices
result['Sum']=result['first']+result['second']
result.head()
```

```python
#I will create a pivot table on sum values and see the sum of probability per each unique sum values
pivot = pd.pivot_table(result,values='probability',index=['Sum'],aggfunc='sum',)
pivotdf = pd.DataFrame(pivot.to_records())
```

```python
pivotdf
```

```python
# To check if this is created correctly, Let's see if the sum of probability is 1
pivotdf['probability'].sum(axis=0)
```

```python
#our return is 0 for all other sums but $50 for 9. 
pivotdf['return']=0
pivotdf.loc[pivotdf['Sum']==9,'return']=50
pivotdf
```

```python
pivotdf['cost']=-20
pivotdf['NetReturn']=pivotdf['return']+pivotdf['cost']
pivotdf['ExpectedReturn']=pivotdf['NetReturn']*pivotdf['probability']
pivotdf
```

```python
# Let's see what is the expected return for each roll of two dices
pivotdf.ExpectedReturn.sum(axis=0)
```

This means that everytime we roll the two dices we are expected to loss 14 dollars.
So **Don't Play This Game**
